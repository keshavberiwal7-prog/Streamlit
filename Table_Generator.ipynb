```python
import streamlit as st

st.title("Table Generator")

number = st.number_input("Enter a number to generate its table:", value = 1, step = 1)

if st.button("Generate Table"):
  st.write(f"Multiplication Table of {number}")

  for i in range(1, 11):
    st.write(f"{number} x {i} = {number * i}")
```
